# ZADANIE 3 (DATASET NIE Z CSV TYLKO IRIS)

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_names'] = df['target'].apply(lambda x: iris.target_names[x])

df.head()
```

```python
from scipy import stats
import numpy as np

# Statystyki opisowe
descriptive_stats = stats.describe(df['sepal length (cm)'])
print("Statystyki opisowe:")
print(descriptive_stats)

# Dominanta
mode_result = stats.mode(df['sepal length (cm)'])
print("\nDominanta:")
print(mode_result)

# Skosność i Kurtoza
skewness = stats.skew(df['sepal length (cm)'])
kurtosis = stats.kurtosis(df['sepal length (cm)'])
print("\nSkosność:", skewness)
print("Kurtoza:", kurtosis)

# Test normalności (test Shapiro-Wilka)
shapiro_test_stat, shapiro_test_p = stats.shapiro(df['sepal length (cm)'])
print("\nTest Shapiro-Wilka:")
print(f"Statystyka testowa: {shapiro_test_stat}, Wartość p: {shapiro_test_p}")

# Przedział ufności dla średniej
confidence_interval = stats.t.interval(0.95, len(df['sepal length (cm)'])-1, loc=np.mean(df['sepal length (cm)']), scale=stats.sem(df['sepal length (cm)']))
print("\n95% Przedział ufności dla średniej:")
print(confidence_interval)

# Test hipotezy dla średniej (test t dla jednej próby)
t_stat, t_p_value = stats.ttest_1samp(df['sepal length (cm)'], popmean=5.0)
print("\nTest t dla jednej próby:")
print(f"Statystyka t: {t_stat}, Wartość p: {t_p_value}")
```

# ZADANIE 4

```python
df = pd.read_csv('../docs/brain_size.csv', sep=';', na_values='.')
df = df.drop(df.columns[0], axis=1)

# Uzupełnianie pustych danych
mean_weight = df['Weight'].mean()
mean_height = df['Height'].mean()
df['Weight'].fillna(mean_weight, inplace=True)
df['Height'].fillna(mean_height, inplace=True)

df
```

```python
print('Mean (VIQ):', df['VIQ'].mean())
print('Male count:', len(df[df['Gender']=='Male']))
print('Female count:', len(df[df['Gender']=='Female']))
```

```python
import matplotlib.pyplot as plt

columns = ['VIQ', 'PIQ', 'FSIQ']
for column in columns:
    plt.hist(df[column], bins=20)
    plt.title(column)
    plt.show()
```

```python
df_female = df[df['Gender']=='Female']
for column in columns:
    plt.hist(df_female[column], bins=20)
    plt.title(f'{column} (female)')
    plt.show()
```
